```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import integrate
from scipy.misc import derivative

import piplite
await piplite.install('ipywidgets')

from ipywidgets import interactive, fixed
```

```python
def cpmm(position_size=1000, pa_t0=40000, pb_t0=1, pa_t1=1, pb_t1=1):
    pa_t1= np.arange(pa_t0*0.9, pa_t0*1.1, 0.01)
    hold_position_value = (position_size*pb_t1)/(2*pb_t0)+(position_size*pa_t1)/(2*pa_t0)
    ratio_change=(pa_t0*pb_t1)/(pb_t0*pa_t1)
    lp_position_value = (2*np.sqrt(ratio_change)*hold_position_value)/(ratio_change+1)
    
    fig, ax = plt.subplots()
    ax.plot(pa_t1, lp_position_value)
    ax.set(xlabel='Price', ylabel='Value', title='LP Position')
    ax.grid()
    fig.savefig('')
    plt.show()
    
    fig2, ax2 = plt.subplots()
    ax2.plot(pa_t1, np.gradient(lp_position_value))
    ax2.set(xlabel='Price', ylabel='Value', title='Delta')
    ax2.grid()
    fig2.savefig('')
    plt.show()
    
    fig3, ax3 = plt.subplots()
    ax3.plot(pa_t1, np.gradient(np.gradient(lp_position_value)))
    ax3.set(xlabel='Price', ylabel='Value', title='Gamma')
    ax3.grid()
    fig3.savefig('')
    plt.show()
    
    fig4, ax4 = plt.subplots()
    ax4.plot(pa_t1, np.gradient(np.gradient(np.gradient(lp_position_value))))
    ax4.set(xlabel='Price', ylabel='Value', title='Speed')
    ax4.grid()
    fig4.savefig('')
    plt.show()

cpmm()
```
